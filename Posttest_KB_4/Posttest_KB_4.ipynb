## Lakukan Data cleaning pada data dengan nilai null

```python
#mengecek data dengan nilai null 
df.isna().sum()
```

```python
#mengecek type
df.info()
```

```python
df.select_dtypes(include=['object']).head(5)
```

```python
#(tulisan : modus)
from sklearn.impute import SimpleImputer
imputerString = SimpleImputer(strategy="most_frequent")
```

```python
df["artist_name"] = imputerString.fit_transform(df[["artist_name"]])
df["track_name"] = imputerString.fit_transform(df[["track_name"]])
df["key"] = imputerString.fit_transform(df[["key"]])
df["mode"] = imputerString.fit_transform(df[["mode"]])
df["tempo"] = imputerString.fit_transform(df[["tempo"]])
df["obtained_date"] = imputerString.fit_transform(df[["obtained_date"]])
df["music_genre"] = imputerString.fit_transform(df[["music_genre"]])
```

```python
df.isna().sum()
```

```python
#mengecek atribut bertipe desimal
df.select_dtypes(include=['float64']).head(5)
```

```python
#(bilangan desimal : mean)
imputerFloat = SimpleImputer(strategy="mean")
```

```python
df['instance_id'] = imputerFloat.fit_transform(df[["instance_id"]])
df['acousticness'] = imputerFloat.fit_transform(df[["acousticness"]])
df['danceability'] = imputerFloat.fit_transform(df[["danceability"]])
df['duration_ms'] = imputerFloat.fit_transform(df[["duration_ms"]])
df['energy'] = imputerFloat.fit_transform(df[["energy"]])
df['instrumentalness'] = imputerFloat.fit_transform(df[["instrumentalness"]])
df['liveness'] = imputerFloat.fit_transform(df[["liveness"]])
df['loudness'] = imputerFloat.fit_transform(df[["loudness"]])
df['speechiness'] = imputerFloat.fit_transform(df[["speechiness"]])
df['valence'] = imputerFloat.fit_transform(df[["valence"]])
```

```python
df.isna().sum()
```

```python
#(bilangan bulat : median)
imputerBulat = SimpleImputer(strategy='median' )
```

```python
df["popularity"] = imputerBulat.fit_transform(df[["popularity"]])

df.isna().sum()
```
