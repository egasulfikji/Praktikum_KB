## PIE CHART

```python
alamat= df[["school","failures"]].groupby(["school"],as_index=False).mean().sort_values(by="failures",ascending=False)
alamat.reset_index(inplace = True)
alamat
```

```python
warna_pie = ("turquoise","teal")
plt.figure(figsize=(10,10))
plt.pie(alamat["failures"],labels=alamat["school"],startangle=90,autopct='%.1f%%', shadow=True, explode=(0.06,0), colors=warna_pie)
plt.title("Persentase Jumlah Kegagalan di Sekolah",color='red',fontsize =17, style="italic")
plt.show()
```

## Kesimpulan = Dari gambar pie chart diatas menghasilkan persentase perbandingan jumlah kegagalan di sekolah MS (Mousinho da Silveira) dengan persentase 58,9% yang paling banyak dan GP (Gabriel Pereira) dengan persentase 41,1% yang paling sedikit

## GRAFIK BATANG

```python
plt.figure(figsize=(10,10)) # panjang 10, lebar 10
sns.countplot(data=df,x= 'reason', palette='icefire', hue='sex')
plt.title('Perbandingan Reason pada gender',color='red',fontsize =17, style="italic")
plt.xlabel('Reason',color='red',fontsize =15)
plt.ylabel('Count',color='red',fontsize =15)
plt.show()
```

## Kesimpulan = Dari grafik diatas menunjukkan bahwa total siswa yang reason/alasan memilih sekolah berdasarkan jenis kelamin

## HEATMAP

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(), annot=True, cmap="Oranges")
plt.title("Korelasi Antar Variabel",color='red',fontsize =17, style="italic")
plt.xlabel('Korelasi X',color='red',fontsize = 15)
plt.ylabel('Korelasi Y',color='red',fontsize = 15)
plt.show()
```

## Kesimpulan =  Grafik HeatMap diatas merupakan visualisasi atau pemetaan dengan menampilkan dari dataset yang sudah di import, yang menjelaskan korelasi antar variabel yang semakin tinggi angka suatu kelompok data maka warnanya akan semakin gelap.

## BOXPLOT

```python
sns.boxplot(x= df['sex'], y=df['age'], palette='rocket')
plt.title("Student Dataset",color='red',fontsize = 17, style="italic")
plt.xlabel('Sex',color='red',fontsize = 15)
plt.ylabel('Age',color='red',fontsize = 15)
plt.show()
```

## Kesimpulan = Boxplot diatas menggambarkan persebaran data dan mencari outliers. dengan X dari data Sex dan Y dari data Age
